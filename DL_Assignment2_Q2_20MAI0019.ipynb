```python
# Import the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Conv2D
import skimage.measure
```

```python
#Importing the image

from PIL import Image
image=Image.open('/content/drive/MyDrive/city1.jpeg')
plt.imshow(image)
```

```python
rsize = image.resize((np.array((15,15))).astype(int))
image = np.asarray(rsize)
image.shape
```

```python
#Puting the image into an array
x=image.reshape(1,15,15,3)
```

```python
def Convolution(x,kernel_size,strides=1, padding='valid'):
 model=Sequential()
 model.add(Conv2D(1, kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(15,15,3)))
 y1 = model.predict(x)
 y=y1.reshape(y1.shape[1],y1.shape[2])
 filtered_image = Image.fromarray(y)
 return filtered_image
```

```python
Filtered_image_1=Convolution(x,(3,3))
Filtered_image_2=Convolution(x,(5,5))
Filtered_image_3=Convolution(x,(3,3),2)
Filtered_image_4=Convolution(x,(5,5),2)
Filtered_image_5=Convolution(x,(3,3),1,'same')
Filtered_image_6=Convolution(x,(5,5),1,'same')
```

```python
# entropy
plt.imshow(Filtered_image_1)
entropy = skimage.measure.shannon_entropy(Filtered_image_1)
print("Entropy:",entropy)
```
